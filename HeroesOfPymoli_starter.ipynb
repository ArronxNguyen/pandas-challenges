```python
# Dependencies and Setup
import pandas as pd

# File
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
heroes_df  = pd.read_csv(file_to_load)
heroes_df.head()
```

## Player Count

* Display the total number of players


```python
#Find uniques IDs
heroes_df["SN"].unique()
# compute the length of unique names 'SN'
heroes_numbers = len(heroes_df["SN"].unique())
#create a new dictionary dataframe to show total players
total_players_df = pd.DataFrame({"Total Players" : [heroes_numbers]})

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# get numerical calculations
unique_items = len(heroes_df['Item Name'].unique())
average_price = heroes_df['Price'].mean()
total_purchases = len(heroes_df["Purchase ID"])
total_revenue = heroes_df['Price'].sum()

#summary table for data
summary_df = pd.DataFrame({"Number of Unique Items" : [unique_items],
                          "Average Price" : [average_price],
                          "Number of Purchases" : [total_purchases],
                          "Total Revenue" : [total_revenue]})
#format clean
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Use numbers by unique IDs first (by SN) not entire list, as same SN can have multiple Purchase IDs
#total_count = heroes_SN_df
#Or
#remove duplicates in SN
no_dups_df = heroes_df.drop_duplicates('SN')
#no_dups_df.count()

#Group by Gender
gender_df = no_dups_df.groupby(['Gender'])
#gender_df.count()

#create two dataframe
# 1st for gender count
gender_df = pd.DataFrame(gender_df['SN'].count())
#Rename SN to Total
gender_df = gender_df.rename(columns=({"SN":"Total"}))
gender_df

#2nd for gender for percentages 
gender_percent_df = pd.DataFrame(gender_df['Total']/sum(gender_df['Total'])) * 100

#Rename Total to Percentage
gender_percent_df = gender_percent_df.rename(columns=({"Total":"Percentage"}))
gender_percent_df

#Merge gender totals and gender percentages dataframe
merge_df = pd.merge(gender_df,gender_percent_df, on='Gender')
merge_df

#clean format at percentage
merge_df["Percentage"] = merge_df["Percentage"].map("{:.2f}%".format)


#sort
merge_df = merge_df.sort_values(by=['Total'], ascending=False)
merge_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Gender
#get count mean sum and nunique
gender_purchase_analysis_groupby_df = heroes_df.groupby(["Gender"]).agg({"Price":["count","mean","sum"],"SN":"nunique"})

#using the .agg function on top of the groupby object, creates a multi-index column and will cause KeyError when refencing for calculations
#either reference the top-level column first before accessing the other columns contained within it
#or flatten the index (Stackoverflow "Pandas - How to flatten a hierarchical index in columns")

#flatten the index
gender_purchase_analysis_groupby_df.columns = [' '.join(col).strip() for col in gender_purchase_analysis_groupby_df.columns.values]

#Rename columns
gender_purchase_analysis_groupby_df = gender_purchase_analysis_groupby_df.rename(columns=({"Price count":"Purchase Count"}))
gender_purchase_analysis_groupby_df = gender_purchase_analysis_groupby_df.rename(columns=({"Price mean":"Average Purchase Price"}))
gender_purchase_analysis_groupby_df = gender_purchase_analysis_groupby_df.rename(columns=({"SN nunique":"Unique Users Count"}))
gender_purchase_analysis_groupby_df = gender_purchase_analysis_groupby_df.rename(columns=({"Price sum":"Total Purchase Value"}))                                                                               

#Add a new column for Average Purchase Total per Person by Gender
gender_purchase_analysis_groupby_df["Average Purchase Total per Person by Gender"] = gender_purchase_analysis_groupby_df['Total Purchase Value']/gender_purchase_analysis_groupby_df['Unique Users Count']

#Formating
gender_purchase_analysis_groupby_df["Average Purchase Price"] = gender_purchase_analysis_groupby_df["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_analysis_groupby_df["Total Purchase Value"] = gender_purchase_analysis_groupby_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_analysis_groupby_df["Average Purchase Total per Person by Gender"] = gender_purchase_analysis_groupby_df["Average Purchase Total per Person by Gender"].map("${:,.2f}".format)

gender_purchase_analysis_groupby_df = pd.DataFrame(gender_purchase_analysis_groupby_df)

#sort
gender_purchase_analysis_groupby_df = gender_purchase_analysis_groupby_df.sort_values(by=['Purchase Count'], ascending=False)
gender_purchase_analysis_groupby_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Figure out the minimum and maximum age 
max_age = heroes_df['Age'].max()
min_age = heroes_df['Age'].min()
min_age
max_age
```

```python
# set up bins for age range
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 80]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


# pd.cut add addtional column to categorise Age into Age Group
heroes_df["Age Group"] = pd.cut(heroes_df["Age"],age_bins, labels=age_group)
#heroes_df

#group data by new column Age Group
heroes_age_df = heroes_df.groupby("Age Group")

#use .aggregate to analyse Age Group & the number of unique SN (nunique) for a total count
heroes_age_df = heroes_age_df.aggregate({"SN":"nunique"})

#rename SN header to Total Count
heroes_age_df = heroes_age_df.rename(columns=({"SN":"Total Count"}))

#get percentage of each Age group count over sum of total count
heroes_age_df["Percentage of Players"] = (heroes_age_df["Total Count"] / sum(heroes_age_df["Total Count"]) * 100)

#format clean
heroes_age_df["Percentage of Players"] = heroes_age_df["Percentage of Players"].map("{:.2f}%".format)

heroes_age_df = pd.DataFrame(heroes_age_df)

heroes_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Use group by Age Group previously defined 
# set up bins for age range
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# pd.cut add addtional column to categorise Age into Age Group
heroes_df["Age Ranges"] = pd.cut(heroes_df["Age"],age_bins, labels=age_ranges)
heroes_df

#get count mean sum and nunique
purchasing_analysis_age_df = heroes_df.groupby(["Age Ranges"]).agg({"SN":["nunique",'count'],"Price":["mean","sum"]})


#flatten the index
purchasing_analysis_age_df.columns = [' '.join(col).strip() for col in purchasing_analysis_age_df.columns.values] 
#purchasing_analysis_age_df.columns = purchasing_analysis_age_df.columns.map(" ".join)

#Rename columns
purchasing_analysis_age_df = purchasing_analysis_age_df.rename(columns=({"SN count":"Purchase Count"}))
purchasing_analysis_age_df = purchasing_analysis_age_df.rename(columns=({"Price mean":"Average Purchase Price"}))
purchasing_analysis_age_df = purchasing_analysis_age_df.rename(columns=({"SN nunique":"Unique Users Count"}))
purchasing_analysis_age_df = purchasing_analysis_age_df.rename(columns=({"Price sum":"Total Purchase Value"})) 

#New column for Avg Total Purchase per Person = Total Purchase Value/ SN
purchasing_analysis_age_df["Average Purchase Total per Person by Age Group"] = purchasing_analysis_age_df["Total Purchase Value"]/purchasing_analysis_age_df["Unique Users Count"]

#formatting
purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_age_df["Total Purchase Value"] = purchasing_analysis_age_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_age_df["Average Purchase Total per Person by Age Group"] = purchasing_analysis_age_df["Average Purchase Total per Person by Age Group"].map("${:,.2f}".format)


purchasing_analysis_age_df = pd.DataFrame(purchasing_analysis_age_df)
purchasing_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by
top_spenders_df = heroes_df.groupby("SN").agg({"Price": ["count","mean","sum"]})

#flatten the index
top_spenders_df.columns = [' '.join(col).strip() for col in top_spenders_df.columns.values]


#Rename columns
top_spenders_df = top_spenders_df.rename(columns=({"Price count":"Purchase Count"}))
top_spenders_df = top_spenders_df.rename(columns=({"Price mean":"Average Purchase Price"}))
top_spenders_df = top_spenders_df.rename(columns=({"Price sum":"Total Purchase Value"}))

#Sort by decending and returning only top 5 Purchase Count
top_spenders_df.sort_values("Total Purchase Value",ascending=False, inplace=True)

#Formarting
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_df = pd.DataFrame(top_spenders_df).head(5)

top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#list df by only relevant data columns
most_popular_items_df = heroes_df[["Item ID","Item Name", "Price"]]

#group by ** Change {} into [] to avoid error**
most_popular_items_df = most_popular_items_df.groupby(["Item ID","Item Name"]).agg(["count","mean","sum"])

#flatten the index
most_popular_items_df.columns = [' '.join(col).strip() for col in most_popular_items_df.columns.values]

#Rename columns
most_popular_items_df = most_popular_items_df.rename(columns=({"Price count":"Purchase Count"}))
most_popular_items_df = most_popular_items_df.rename(columns=({"Price mean":"Item Price"}))
most_popular_items_df = most_popular_items_df.rename(columns=({"Price sum":"Total Purchase Value"}))

#Sort by decending
most_popular_items_df.sort_values("Total Purchase Value",ascending=False, inplace=True)

#Formarting
most_popular_items_df["Item Price"] = most_popular_items_df["Item Price"].map("${:,.2f}".format)
most_popular_items_df["Total Purchase Value"] = most_popular_items_df["Total Purchase Value"].map("${:,.2f}".format)


most_popular_items_df = pd.DataFrame(most_popular_items_df).head(5)
most_popular_items_df
```
